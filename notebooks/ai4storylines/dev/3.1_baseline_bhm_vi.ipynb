## Baseline Model - Linear Regression

#### Input Data

```python
covariate_names = ["sst", "sm_sur", "t2m", "z500_zonal"]
qoi_names = ["tasmax"]
x = df.sort_values(by=["model"])[covariate_names].values
u = ds.sortby("model").tasmax.values
```

We see that we have the `4` covariates and the `19x21` grid points.

Linear Regression only works with a 1D vector. 
So we need to flatten the `u` variable.

```python
x_shape = x.shape
u_shape = u.shape

u_flattened = einops.rearrange(u, "Model Lat Lon -> Model (Lat Lon)")
```

```python
u_shape, u_flattened.shape
```

#### Model

```python
num_spatial = u_flattened.shape[1]
num_models = u_flattened.shape[0]
num_covariates = x.shape[1]
```

```python
u_flattened.shape
```

```python
def model(x: Float[Array, "M D"], y: Float[Array, "M N"]=None):
    num_models, num_features = x.shape
    # noise
    scale = numpyro.sample("scale", dist.LogNormal(0.0, 10.0))
    z_scale: Float[Array, "N"] = numpyro.sample("z_scale", dist.LogNormal(0.0, 10.0))
    # priors on mean
    loc_prior_mean = numpyro.sample("loc_prior_mean", dist.Normal(0.0, 10.))
    loc_prior_scale = numpyro.sample("loc_prior_scale", dist.LogNormal(0.0, 10.))
    # priors on bias
    bias_prior_mean = numpyro.sample("bias_prior_mean", dist.Normal(0.0, 10.))
    bias_prior_scale = numpyro.sample("bias_prior_scale", dist.LogNormal(0.0, 10.))
    # spatial locations
    with numpyro.plate("spatial", num_spatial):
        # sample from bias (N,)
        bias: Float[Array, "N"] = numpyro.sample("bias", dist.Normal(bias_prior_mean, bias_prior_scale))
        
        
        with numpyro.plate("covariates", num_covariates):
            # sample weights, (N,D)
            weight: Float[Array, "D N"] = numpyro.sample("weight", dist.Normal(loc_prior_mean, loc_prior_scale))
        
    # conditional latent variable, (M,N)
    z: Float[Array, "M N"] = jnp.einsum("DN,MD->MN",weight, x) + bias + z_scale
    # Data Likelihood
    numpyro.sample("obs", dist.Normal(z, scale), obs=y)
```

```python
prior_predictive = Predictive(model, num_samples=10)
prior_samples = prior_predictive(jrandom.PRNGKey(0), x=x)
# prior_samples
```

```python
prior_samples["bias"].shape
prior_samples["weight"].shape
```

```python
weights = einops.rearrange(prior_samples["weight"][0], "Variable (Lat Lon) -> Variable Lat Lon", Lat=u.shape[1], Lon=u.shape[2])
```

### Variational Distribution

```python
from numpyro.infer.autoguide import AutoDelta, AutoMultivariateNormal

guide = AutoMultivariateNormal(model)
```

### Inference

$$
\boldsymbol{\phi}^* = 
\underset{\boldsymbol{\phi}}{\text{argmin}}
\hspace{2mm}
\text{ELBO}_{q}
\left[q(\boldsymbol{\theta};\boldsymbol{\phi}), p(y,\boldsymbol{\theta})\right]
$$

```python
from numpyro.infer import Predictive, SVI, Trace_ELBO
optimizer = numpyro.optim.Adam(step_size=0.001)
svi = SVI(model, guide, optimizer, loss=Trace_ELBO())
```

```python
num_steps = 2_000
rng_key = jrandom.PRNGKey(123)
rng_key_train, rng_key = jrandom.split(rng_key, num=2)

svi_result = svi.run(rng_key=rng_key_train, num_steps=num_steps, x=x, y=u_flattened)
```

### Training

```python
fig, ax = plt.subplots()
ax.plot(svi_result.losses)
plt.tight_layout()
plt.show()
```

### Posterior Samples

```python
params = svi_result.params
# samples = guide.get_posterior(params)
predictive = Predictive(guide, params=params, num_samples=1000)
posterior_samples = predictive(jrandom.PRNGKey(0), x=x)
# posterior_samples
```
